## Decision Tree Regressor Model

```python
data = df[['car_name', 'yr_mfr', 'fuel_type', 'kms_run', 'body_type', 'transmission', 'make', 'model', 'car_rating', 'sale_price']]
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()
```

```python
data.loc[:,'car_name'] = le.fit_transform(data['car_name'])
```

```python
data.loc[:,'fuel_type'] = le.fit_transform(data['fuel_type'])
data.loc[:,'body_type'] = le.fit_transform(data['body_type'])
data.loc[:,'transmission'] = le.fit_transform(data['transmission'])
data.loc[:,'make'] = le.fit_transform(data['make'])
data.loc[:,'model'] = le.fit_transform(data['model'])
data.loc[:,'car_rating'] = le.fit_transform(data['car_rating'])
```

```python
X = data.drop(columns='sale_price')
y = data['sale_price']
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.2, random_state=42)
```

```python
dec = DecisionTreeRegressor(random_state=42)
```

```python
params = {
    'max_depth': [2,3,5,7,10, None],
    'min_samples_leaf': [5,10,15,10,25],
    'min_samples_split': [2,5,10,15,20],
    'max_features': ['sqrt','log2', None],
    'ccp_alpha': [0.001,0.01,0.1,0.0,1],
    'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson']
}
```

```python
model = RandomizedSearchCV(estimator=dec,
                           param_distributions=params,
                           n_iter=10,
                           verbose=1,
                           n_jobs=1,
                           random_state=42,
                           scoring=r2_score)
```

```python
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
best_model = model.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

#### Untunned Decision tree regressor Model

```python
model = DecisionTreeRegressor(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

## Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np 
```

```python
for_reg = RandomForestRegressor()
```

```python
vals = np.arange(1,51)
val = vals.tolist()
max_val = val + [None]
parameters = {
    'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
    'max_depth': max_val,
    'min_samples_split': val,
    'min_samples_leaf': val,
    'max_features': ['sqrt', 'log2'],
    'ccp_alpha': [0.0001,0.001,0.01,0.1,0.0,1,10,100,1000]
}
```

```python
tuned = RandomizedSearchCV(estimator=for_reg,
                           param_distributions=parameters,
                           verbose=1,
                           n_jobs=2,
                           n_iter=10,
                           random_state=42,
                           scoring=r2_score)
```

```python
tuned.fit(X_train, y_train)
```

```python
tuned.best_params_
```

```python
best_model = tuned.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
model = RandomForestRegressor(random_state=42)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
